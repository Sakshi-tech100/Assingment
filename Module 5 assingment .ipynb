```python
Q1) Discuss string slicing and provide examples?
Ans- String slicing in Python is a powerful feature that allows you to extract a portion of a string by specifying a start and end index. 
The syntax for slicing a string is-
python
string[start:end:step]

1. Example 
   python
   my_string = "Hello, World!"
   sliced = my_string[0:5]  # Returns 'Hello'
   
2. Example 
   python
   my_string = "Hello, World!"
   sliced = my_string[::2]  # Returns 'Hlo ol!'

3. Example 
   python
   my_string = "Hello, World!"
   reversed_string = my_string[::-1]  # Returns '!dlroW ,olleH'
  ' '
Q2) Explain the key feature list in python?
Ans- Some key features of lists in Python:
-Ordered: Lists maintain the order of elements. The items are indexed, starting from 0, which means you can access them using their position.
-Mutable: Lists are mutable, meaning you can change, add, or remove elements after the list has been created.
-Heterogeneous: Lists can contain items of different data types. For example, you can have integers, strings, and even other lists all in the same list.
-Dynamic Size: Lists can grow or shrink in size as you add or remove elements. You don’t need to declare the size of a list when you create it.
-Support for Nested Lists: Lists can contain other lists as elements, allowing for multi-dimensional data structures.

Q3) Describe how to access, Modify, and delete elementsin a list with examples?
Ans- Accessing, modifying, and deleting elements in a list in Python is straightforward
1. Accessing Elements:
You can access elements in a list using their index. Remember that indexing starts at 0.

Example-
my_list = [10, 20, 30, 40, 50]
first_element = my_list[0]  # Accesses the first element (10)
print(first_element)  # Output: 10

Modifying Elements:
To modify an element, you can assign a new value to a specific index in the list.
my_list = [10, 20, 30, 40, 50]
my_list[2] = 35  # Changes the element at index 2 (30) to 35
print(my_list)  # Output: [10, 20, 35, 40, 50]

Deleting Elements:
You can delete elements using the del statement, the remove() method, or the pop() method.
my_list = [10, 20, 30, 40, 50]
del my_list[1]  # Deletes the element at index 1 (20)
print(my_list)  # Output: [10, 30, 40, 50]


Q4) Compare and contrast tuple and list with example?
Ans- Lists are mutable,	Tuples are immutable
The implication of iterations is Time-consuming, the implication of iterations is comparatively Faster
The list is better for performing operations, such as insertion and deletion. A Tuple data type is appropriate for accessing the elements
Lists consume more memory, Tuple consumes less memory as compared to the list
Lists have several built-in methods, Tuple does not have many built-in methods.
Unexpected changes and errors are more likely to occur	Because tuples don’t change they are far less error-prone.
Example of list- my_list = [item1, item2, item3, ...]
Example of tuple- my_tuple = (item1, item2, item3, ...)


Q5) Describe key feature of set and provide example for their use?
Ans- Sets are a built-in data structure in Python that store unordered collections of unique items. 
    here are some key features of sets, along with examples of their use:
-Unordered:The elements in a set do not have a specific order. When you iterate through a set, the order of elements may vary.
-Unique Elements:Sets automatically enforce uniqueness, meaning that duplicate values are not allowed. If you try to add a duplicate, it will simply be ignored.
-Mutable:You can add or remove elements from a set after it has been created, but the elements themselves must be immutable (for example, you can’t use lists or dictionaries as set elements).
-Fast Membership Testing Sets provide efficient operations for checking whether an item is in the set, which is generally faster than lists.
Examples of set 
Creating a Set:
   my_set = {1, 2, 3, 4}
   print(my_set)  # Output: {1, 2, 3, 4}
   
Adding and Removing Elements:
   my_set.add(5)  # Adding an element
   print(my_set)  # Output: {1, 2, 3, 4, 5}
   
   my_set.remove(3)  # Removing an element
   print(my_set)  # Output: {1, 2, 4, 5}
   
Handling Duplicates:
   my_set = {1, 2, 2, 3}  # Duplicates will be ignored
   print(my_set)  # Output: {1, 2, 3}


Set Operations:
   set_a = {1, 2, 3}
   set_b = {3, 4, 5}
   
   union = set_a | set_b  # Union
   print(union)  # Output: {1, 2, 3, 4, 5}
   
   intersection = set_a & set_b  # Intersection
   print(intersection)  # Output: {3}
   
   difference = set_a - set_b  # Difference
   print(difference)  # Output: {1, 2}


Q6) Discuss the use case of tuple and set in python programming?
Ans- Tuples Definition: A tuple is an immutable, ordered collection of items. Once a tuple is created, you cannot change its elements.
Use Cases-
Data Integrity: Since tuples are immutable, they are often used to store data that should not be changed. This is particularly useful for constants or fixed collections of data.

   Example: Storing the coordinates of a point in a 2D space:
   python
   point = (3, 5)  # Immutable coordinates
Return Multiple Values: Tuples are often used to return multiple values from a function. This allows you to pack multiple results together.

   Example: A function that returns the minimum and maximum values from a list:
   def min_max(numbers):
       return (min(numbers), max(numbers))
   
   result = min_max([1, 2, 3, 4, 5])
   print(result)  # Output: (1, 5)
Dictionary Keys: Tuples can be used as keys in dictionaries because they are hashable (unlike lists). This is useful for composite keys.

   Example: Using tuples as keys in a dictionary to represent a grid:
   python
   grid = {}
   grid[(0, 0)] = "Start"
   grid[(1, 0)] = "End"
   

Sets Definition: A set is an unordered collection of unique items. Sets are mutable, meaning you can add or remove items.
Use Cases:
Removing Duplicates: Sets are ideal for removing duplicates from a list, as they automatically enforce uniqueness.

   Example:
   my_list = [1, 2, 2, 3, 4, 4, 5]
   unique_items = set(my_list)
   print(unique_items)  # Output: {1, 2, 3, 4, 5}
   
Membership Testing: Sets provide fast membership testing, which can be useful in scenarios where you need to check if an item exists in a collection frequently.

   Example:
   my_set = {1, 2, 3, 4, 5}
   print(3 in my_set)  # Output: True

Mathematical Operations: Sets support operations like union, intersection, and difference, making them useful for mathematical computations.

   Example:
   set_a = {1, 2, 3}
   set_b = {2, 3, 4}
   intersection = set_a & set_b  # Output: {2, 3}


Q7) Describe how to add, delete, modify items in dictionary with examples?
Ans- In Python, dictionaries are mutable data structures that store data in key-value pairs. You can easily add, delete, and modify items in a dictionary. Here’s how you can do each of these operations with examples:
Adding Items
To add an item to a dictionary, you can simply assign a value to a new key. If the key already exists, it will update the value for that key.
Example:
# Creating a dictionary
my_dict = {'name': 'Alice', 'age': 25}

# Adding a new item
my_dict['city'] = 'New York'

print(my_dict)
# Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}

Modifying Items
To modify an existing item in a dictionary, you can assign a new value to an existing key.

Example:
Existing dictionary
my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}

# Modifying an existing item
my_dict['age'] = 26

print(my_dict)
# Output: {'name': 'Alice', 'age': 26, 'city': 'New York'}

Deleting Items
To delete an item from a dictionary, you can use the del statement or the pop() method. The del statement removes the key-value pair, while pop() removes the item and returns its value.

    Example using del-
# Existing dictionary
my_dict = {'name': 'Alice', 'age': 26, 'city': 'New York'}

# Deleting an item
del my_dict['city']
print(my_dict)
# Output: {'name': 'Alice', 'age': 26}

Q8) Discuss the importance of dictionary keys being inmutable and provide examples?
Ans- In Python, dictionary keys must be immutable types, such as strings, numbers, or tuples. This immutability is crucial for several reasons:
Hashing
Dictionaries in Python use a hash table to store key-value pairs. When you add a key to a dictionary, Python computes a hash value based on the key. 
 This hash value is used to quickly locate the corresponding value. If the key were mutable (like a list), changing the key would change its hash value,
 making it impossible for Python to find the value associated with that key.
Example:
# Immutable key (string)
my_dict = {'name': 'Alice'}
print(my_dict['name'])  # Output: Alice

# Mutable key (list) - This will raise an error
# my_dict = {[1, 2, 3]: 'value'}  # TypeError: unhashable type: 'list'

Consistency
Having immutable keys ensures that the dictionary remains consistent. Once a key is added, it cannot change, which means that the relationship between the key and its value remains stable throughout the dictionary's lifetime. This consistency is essential for data integrity.
Example:
# Example of consistent key-value pairing
my_dict = {1: 'One', 2: 'Two'}
print(my_dict[1])  # Output: One

# If keys were mutable, changing them would break this consistency
Performance
Using immutable keys allows for faster access and retrieval. Since the hash value of an immutable key does not change, 
 Python can quickly compute the location of the corresponding value without recalculating or checking for changes.









   









    



```
