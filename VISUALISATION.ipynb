```python
#2 Using the Student Grades, create a violin plot to display the distribution of scores across different grade
#categories.?

import matplotlib.pyplot as plt
import numpy as np
 
# Generate random data
np.random.seed(1)
data = np.random.randn(100)
 
# Create a violin plot
plt.figure()
plt.violinplot(data, showmedians=True)
 
# Set plot labels and title
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Univariate Violin Plot')
 
# Show the plot
plt.show()
```

```python
#4 Using the given x and y data, generate a 3D surface plot to visualize the function ?

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
 
ax.set_title('3D Surface plot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
#2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns.?

from bokeh.io import curdoc
from bokeh.plotting import figure, output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 6, 4, 5]

output_file("dark_minimal.html")

curdoc().theme = 'dark_minimal'

p = figure(title='dark_minimal', width=300, height=300)
p.line(x, y)

show(p)
```

```python
#3  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.?

from bokeh.plotting import figure, show

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
counts = [5, 3, 4, 2, 4, 6]

p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0

show(p)
```

```python
#4 Create a Bokeh histogram to visualize the distribution of the given data.
import numpy as np
from bokeh.plotting import figure, show, output_notebook

output_notebook()

# synthesize example data
measured = np.random.normal(0, 0.5, 1000)

hist, edges = np.histogram(measured, density=True, bins=50)

p = figure(title="Normal Distribution (μ=0, σ=0.5)")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], line_color=None)

show(p)
```
